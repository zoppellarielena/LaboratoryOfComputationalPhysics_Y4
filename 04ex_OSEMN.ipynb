1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
import random as rnd
l = rnd.sample(range(0, 100000), 100)
s=[]
for i in l:
    s.append(str(i))
    s.append(', ')
a=open("simple_data.txt",'w')
for i in s:
    a.write(i)
a.close()
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
import numpy as np
import numpy.random as npr

matrix=npr.rand(5,5)
print(matrix)
r=[]
for i in matrix:
    for j in i:
        r.append(str(j))
        r.append(" ")
    r.append('\n')
b=open("data.txt",'w')
for i in r:
    b.write(i)
b.close()
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
file=open("data.txt")
new=open("data1.cvs",'w')
for line in file:
    for c in line.split(' '):
        new.write(c)
        if c!="\n":
            new.write(",")
new.close()
```
